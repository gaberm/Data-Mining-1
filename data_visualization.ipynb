```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("preprocessed_survey_malte.csv", sep=";")
data.head()
```

```python
highest_degree = np.arange(8)
width = 0.2
fig, ax = plt.subplots()
fig.set_figwidth(15)
ax.bar(highest_degree - width, data.groupby("w1_highest_degree").mean()["w1_satisfaction_democracy"]/4, width, label="wave1")
ax.bar(highest_degree, data.groupby("w2_highest_degree").mean()["w2_satisfaction_democracy"]/4, width, label="wave2")
ax.bar(highest_degree + width, data.groupby("w3_highest_degree").mean()["w3_satisfaction_democracy"]/4, width, label="wave3")
ax.set_xlabel("highest degree")
ax.set_ylabel("satisfaction with democracy")
ax.set_ybound(0.3, 0.7)
ax.legend()
```

```python
political_interest = np.arange(5)
width = 0.2
fig, ax = plt.subplots()
fig.set_figwidth(15)
ax.bar(political_interest - width, data.groupby("w1_polint").mean()["w1_satisfaction_democracy"]/4, width, label="wave1")
ax.bar(political_interest, data.groupby("w2_polint").mean()["w2_satisfaction_democracy"]/4, width, label="wave2")
ax.bar(political_interest + width, data.groupby("w3_polint").mean()["w3_satisfaction_democracy"]/4, width, label="wave3")
ax.set_xlabel("highest degree")
ax.set_ylabel("satisfaction with democracy")
ax.set_ybound(0.3, 0.7)
ax.legend()
```

```python
w1_pid_par = data[["w1_pid1_par_spd", 
                  "w1_pid1_par_cducsu", 
                  "w1_pid1_par_gruene",
                  "w1_pid1_par_fdp",
                  "w1_pid1_par_afd",
                  "w1_pid1_par_linke",
                  "w1_pid1_par_sonstige"]]
w2_pid_par = data[["w2_pid1_par_spd", 
                  "w2_pid1_par_cducsu", 
                  "w2_pid1_par_gruene",
                  "w2_pid1_par_fdp",
                  "w2_pid1_par_afd",
                  "w2_pid1_par_linke",
                  "w2_pid1_par_sonstige"]]
w3_pid_par = data[["w3_pid1_par_spd", 
                  "w3_pid1_par_cducsu", 
                  "w3_pid1_par_gruene",
                  "w3_pid1_par_fdp",
                  "w3_pid1_par_afd",
                  "w3_pid1_par_linke",
                  "w3_pid1_par_sonstige"]]
w1_nid_par = data[["w1_nid1_par_spd", 
                  "w1_nid1_par_cducsu", 
                  "w1_nid1_par_gruene",
                  "w1_nid1_par_fdp",
                  "w1_nid1_par_afd",
                  "w1_nid1_par_linke",
                  "w1_nid1_par_sonstige"]]
w2_nid_par = data[["w2_nid1_par_spd", 
                  "w2_nid1_par_cducsu", 
                  "w2_nid1_par_gruene",
                  "w2_nid1_par_fdp",
                  "w2_nid1_par_afd",
                  "w2_nid1_par_linke",
                  "w2_nid1_par_sonstige"]]
w3_nid_par = data[["w3_nid1_par_spd", 
                  "w3_nid1_par_cducsu", 
                  "w3_nid1_par_gruene",
                  "w3_nid1_par_fdp",
                  "w3_nid1_par_afd",
                  "w3_nid1_par_linke",
                  "w3_nid1_par_sonstige"]]
pid_nid_parties = [w1_pid_par,
                   w2_pid_par,
                   w3_pid_par,
                   w1_nid_par,
                   w2_nid_par,
                   w3_nid_par]

w1_party_democracy = pd.concat([w1_pid_par.idxmax(axis=1).reset_index(), data["w1_satisfaction_democracy"]], axis=1)
w1_party_democracy = w1_party_democracy.rename(columns={0: "party"})
w2_party_democracy = pd.concat([w2_pid_par.idxmax(axis=1).reset_index(), data["w2_satisfaction_democracy"]], axis=1)
w2_party_democracy = w2_party_democracy.rename(columns={0: "party"})
w3_party_democracy = pd.concat([w3_pid_par.idxmax(axis=1).reset_index(), data["w3_satisfaction_democracy"]], axis=1)
w3_party_democracy = w3_party_democracy.rename(columns={0: "party"})

w1_nparty_democracy = pd.concat([w1_nid_par.idxmax(axis=1).reset_index(), data["w1_satisfaction_democracy"]], axis=1)
w1_nparty_democracy = w1_nparty_democracy.rename(columns={0: "party"})
w2_nparty_democracy = pd.concat([w2_nid_par.idxmax(axis=1).reset_index(), data["w2_satisfaction_democracy"]], axis=1)
w2_nparty_democracy = w2_nparty_democracy.rename(columns={0: "party"})
w3_nparty_democracy = pd.concat([w3_nid_par.idxmax(axis=1).reset_index(), data["w3_satisfaction_democracy"]], axis=1)
w3_nparty_democracy = w3_nparty_democracy.rename(columns={0: "party"})
```

```python
labels = ["AfD", "CDU/CSU", "FDP", "BÃ¼ndnis 90/Die GrÃ¼nen", "Die Linke", "Sonstige", "SPD"]
party = np.arange(7)
width = 0.2
fig, ax = plt.subplots(1,1)
fig.set_figwidth(15)

ax.bar(party - width, w1_party_democracy.groupby("party").mean()["w1_satisfaction_democracy"]/4*100, width, label="wave1")
ax.bar(party, w2_party_democracy.groupby("party").mean()["w2_satisfaction_democracy"]/4*100, width, label="wave2")
ax.bar(party + width, w3_party_democracy.groupby("party").mean()["w3_satisfaction_democracy"]/4*100, width, label="wave3")

ax.set_xticks(party)
ax.set_xticklabels(labels)
ax.set_ybound(30, 70)
ax.set_xlabel("Affiliated party")
ax.set_ylabel("satisfaction with democracy in percent")
ax.legend()
```
